```python
import cv2
import matplotlib.pyplot as plt
from IPython.display import Image
import numpy as np
```

```python
ch3_bgr=cv2.imread("chandrayan3.webp",cv2.IMREAD_COLOR)
ch3_rgb=ch3_bgr[:,:,::-1]
```

```python
plt.imshow(ch3_rgb)
```

```python
image=ch3_rgb.copy()

cv2.line(image,(1000,500),(2000,500),(0,255,0),thickness=50, lineType=cv2.LINE_AA)
cv2.line(image,(1700,300),(2000,500),(0,255,0),thickness=50, lineType=cv2.LINE_AA)
cv2.line(image,(1700,700),(2000,500),(0,255,0),thickness=50, lineType=cv2.LINE_AA)
plt.imshow(image)
```

```python
image=ch3_rgb.copy()
cv2.circle(image,(2470,400),350,(255,0,0),thickness=50,lineType=cv2.LINE_AA)
plt.imshow(image)
```

```python
image=ch3_rgb.copy()
cv2.putText(image,"Chandrayan-3",(1000,3100),cv2.FONT_HERSHEY_PLAIN,25,(100,255,0),thickness=50,lineType=cv2.LINE_AA)
plt.imshow(image)
```
