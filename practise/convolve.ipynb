```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

```python
img1=cv2.imread('C:\\Users\\kusha\\Desktop\\CEVI-Summer_School\\images\\checkerboard.png')
img2=img1[:,:,1]
plt.imshow(img2,cmap='gray')
```

```python
gaussian = (1 / 16.0) * np.array([[1., 2., 1.],
                                  [2., 4., 2.],
                                  [1., 2., 1.]])
```

```python
kernel=np.ones((10,10),np.float32)*1/9
result=scipy.signal.convolve2d(img2, gaussian)
```

```python
def multi_convolver(image, kernel, iterations):
    for i in range(iterations):
        image = scipy.signal.convolve2d(image, kernel, 'same', boundary = 'fill',
                           fillvalue = 0)
    return image
```

```python
result=(multi_convolver(img2, kernel,15))
```

```python
plt.imshow(result,cmap='gray')
```
