```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

def read_image(path):
    return cv2.imread(path, cv2.IMREAD_COLOR)
def convert_to_hsv(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
def modify_value(image, theta):
    image[:,:,2]=np.clip(image[:,:,2]+theta,0,254).astype(np.uint8)
    return image
def modify_saturation(image,eps,a=-3.2,c=0.4):
    image[:,:,1]=(np.exp(np.log(image[:,:,1]**(1/c)) - np.log((1 - image[:,:,1])**(1/c)) + a * eps))/1 + np.exp(np.log(image[:,:,1]**(1/c)) - np.log((1 - image[:,:,1])**(1/c)) + a * eps)
    image[:,:,1]=np.clip(image[:,:,1],0,255).astype(np.uint8)
    return image
```

```python
img_org= cv2.imread("./balloon.jpg",cv2.IMREAD_COLOR)
plt.imshow(img_org[:,:,::-1])
```

```python
img_org_hsv=cv2.cvtColor(img_org,cv2.COLOR_BGR2HSV)
img_org_hsv=modify_value(img_org_hsv,3)
img_org_hsv=modify_saturation(img_org_hsv,1)
```
